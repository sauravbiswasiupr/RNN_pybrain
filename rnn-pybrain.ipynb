```python
from pybrain.structure import RecurrentNetwork, LinearLayer,SigmoidLayer,FullConnection,SoftmaxLayer
from pybrain.tests.helpers import gradientCheck
n =  RecurrentNetwork()
n.addInputModule(LinearLayer(1, name='in'))
n.addModule(SigmoidLayer(5, name='hidden'))
n.addOutputModule(SoftmaxLayer(2, name='out'))
n.addConnection(FullConnection(n['in'], n['hidden'], name='c1'))
n.addConnection(FullConnection(n['hidden'], n['out'], name='c2'))
n.addRecurrentConnection(FullConnection(n['hidden'], n['hidden'], name='c3'))

n.sortModules()
gradientCheck(n)
```

```python
n.activate((2))
```

```python
#define the input as a sine wave 
x=linspace(-100,100)
y=sin(x)
plot(y)
pred=[]
for i in y :
    pred.append(n.activate((i)))
pred = array(pred)
plot(pred)
```

```python
#let's write a piece of code that creates different sine waves , based on different frequences but on the same x values i.e sin(wx) , where w is freq 
w = [i for i in range(1,21)]
x =linspace(-100,100)
sin_waves=[]
for omega  in w : 
    wave=sin(omega*x)
    sin_waves.append(wave)
sin_waves=array(sin_waves)
for wave in sin_waves[:5]:
    plot(wave)
print "Sample seq : ", sin_waves[0] , 
print sin_waves.shape 
```

```python
#create the dataset 
from pybrain.datasets import SequenceClassificationDataSet
DS = SequenceClassificationDataSet(1,1,nb_classes=2) 
#there are 20 sequences and each seq has 50 timesteps 
for i in xrange(10):
    #initialize a new sequence 
    DS.newSequence()
    #start adding timesteps for each seq 
    for j in xrange(50):
        DS.addSample([sin_waves[i][j]],[0])
for i  in xrange(10,20):
    DS.newSequence()
    for j in xrange(50):
       DS.addSample([sin_waves[i][j]],[1])
```

```python
trndata = DS
trndata._convertToOneOfMany(bounds=[0.,1.])
print trndata.indim 

inp = trndata['input']
plot(inp[:,:])
plot(trndata['target'][:,0],'-o')
```

```python
#now define the trainer and train on the dataset
from pybrain.supervised import BackpropTrainer 
trainer = BackpropTrainer(n,dataset=trndata,verbose=True)
trainer.trainUntilConvergence(maxEpochs=100)
#So I silenced the training , because we already saw it earlier 
```
