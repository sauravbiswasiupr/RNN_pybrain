```python
#notebook to test out a simple rnn architecture 
from pybrain.structure import LinearLayer,SigmoidLayer,FullConnection,TanhLayer
from pybrain.structure.modules import SoftmaxLayer
from pybrain.structure.networks import RecurrentNetwork
from numpy import * 
from pylab import * 
from matplotlib import pyplot as plt 
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.structure.modules import TanhLayer

rnn = RecurrentNetwork() 
inLayer=LinearLayer(1) 
hiddenLayer = TanhLayer(5)
outLayer=LinearLayer(1) 
rnn.addInputModule(inLayer)
rnn.addModule(hiddenLayer)
rnn.addOutputModule(outLayer)
#connection wts 
in_to_hidden = FullConnection(inLayer,hiddenLayer)
hidden_to_hidden=FullConnection(hiddenLayer,hiddenLayer)
hidden_to_out = FullConnection(hiddenLayer,outLayer)
rnn.addConnection(in_to_hidden)
rnn.addRecurrentConnection(hidden_to_hidden)
rnn.addConnection(hidden_to_out)
rnn.sortModules()
print rnn
```

```python
#create a sample dataset 
#of y=sin (ax) + cos (bx)
#sample activation 
x=linspace(-20*pi,20*pi) 
y = cos(x)
act=[] 
for i in range(len(y)):
    act.append(rnn.activate((y[i])))

plot(y,'r') 

plot(act,'g')
```

```python
#now we make a SequenceClassificationDataSet and see if we can fit in MNIST values there and train our rnn for that purpose
#import the MNIST dataset first 
import cPickle 
f =  open('/home/saurav/Desktop/CVPR_work/ocropus/ocropy/mnist.pkl','rb')
train,valid,test=cPickle.load(f) 
train_x,train_y=train
valid_x,valid_y = valid 
test_x,test_y=test
from pybrain.datasets.classification import SequenceClassificationDataSet 
```

```python
def create_dataset(inp,targ,length):
   ds = SequenceClassificationDataSet(784,1,nb_classes=10) 
   for i in range(length): 
        ds.newSequence()
        img=inp[i]
        target=targ[i] 
        ds.addSample(img,target)
        print "Sequence length now : " , ds['input'].shape
                                        
   return ds  
```

```python
train_ds=create_dataset(train_x[:10],train_y[:10],train_x[:10].shape[0])
valid_ds=create_dataset(valid_x[:10],valid_y[:10],valid_x[:10].shape[0])
test_ds=create_dataset(test_x[:10],test_y[:10],test_x[:10].shape[0])
train_ds._convertToOneOfMany()
valid_ds._convertToOneOfMany()
test_ds._convertToOneOfMany()
```

```python
#now let us train data on the toy RNN
from pybrain.utilities import percentError
from pybrain.supervised.trainers import BackpropTrainer
```

```python
trainer=BackpropTrainer(rnn,dataset=train_ds,momentum=0.1,verbose=True)
maxEpochs=50
trnerror=[]
for epoch in range(maxEpochs):
    trainer.trainEpochs(1)
    training_error=percentError(trainer.testOnClassData(),train_ds['class'])
    trnerror.append(training_error)
```
